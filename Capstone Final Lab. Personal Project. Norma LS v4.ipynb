```python
pc_district = full_df.groupby(['Postcode district','District'],as_index=False)['County'].count()
```

```python
#Setting the common col as index to concatenate
pc_ward = pc_ward.set_index(['Postcode district'])
pc_district = pc_district.set_index(['Postcode district'])
```

```python
# Now concatenating using inner join and reseting index
london_df = pd.concat([ pc_district,pc_ward], axis=1, join='inner').reset_index()
```

```python
london_df.drop(['County'], axis=1, inplace=True)
```
